```python
import pandas as pd

# 데이터 불러오기
def load_data():
    congestion_data = pd.read_csv('서울교통공사_지하철혼잡도정보_20240930.csv',encoding='cp949')
    ridership_data = pd.read_csv('서울시 지하철호선별 역별 승하차 인원 정보.csv', encoding='cp949')
    return congestion_data, ridership_data

# 데이터 로드
congestion_data, ridership_data = load_data()
print(congestion_data)
print(ridership_data)
```

```python
# 8호선 데이터 필터링
def filter_data(congestion_data, ridership_data):
    congestion_8 = congestion_data[congestion_data['호선'] == 8]
    ridership_8 = ridership_data[ridership_data['호선명'] == '8호선']
    return congestion_8, ridership_8

# 8호선 필터링
congestion_8, ridership_8 = filter_data(congestion_data, ridership_data)
print(congestion_8)
print(ridership_8)
```

```python
# 각 컬럼별 결측치 개수 확인
print("혼잡도 데이터 결측치 개수:")
print(congestion_8.isnull().sum())

print("\n승하차 데이터 결측치 개수:")
print(ridership_8.isnull().sum())
```

```python
# 요일별 결측치 확인 함수
def check_missing_values_by_day(congestion_8):
    # 요일별로 데이터 분리
    saturday_data = congestion_8[congestion_8['요일구분'] == '토요일']
    sunday_data = congestion_8[congestion_8['요일구분'] == '일요일']
    weekday_data = congestion_8[congestion_8['요일구분'] == '평일']
    
    # 결측치 계산
    saturday_missing = saturday_data.isnull().sum()
    sunday_missing = sunday_data.isnull().sum()
    weekday_missing = weekday_data.isnull().sum()
    
    # 결과 출력
    print("토요일 결측치:\n", saturday_missing)
    print("\n일요일 결측치:\n", sunday_missing)
    print("\n평일 결측치:\n", weekday_missing)
    
    # 결과 반환
    return saturday_missing, sunday_missing, weekday_missing

# 함수 실행
saturday_missing, sunday_missing, weekday_missing = check_missing_values_by_day(congestion_8)
```

```python
# 결측치 제거
congestion_8_cleaned = congestion_8.dropna()

# 요일별 결측치 확인 함수
def check_missing_values_by_day(cleaned_data):
    # 요일별로 데이터 분리
    saturday_data = cleaned_data[cleaned_data['요일구분'] == '토요일']
    sunday_data = cleaned_data[cleaned_data['요일구분'] == '일요일']
    weekday_data = cleaned_data[cleaned_data['요일구분'] == '평일']
    
   # 결측치 계산 후 결과를 정수로 변환
    saturday_missing = saturday_data.isnull().sum().astype(int)
    sunday_missing = sunday_data.isnull().sum().astype(int)
    weekday_missing = weekday_data.isnull().sum().astype(int)

    # 결과 출력
    print("토요일 결측치:\n", saturday_missing)
    print("\n일요일 결측치:\n", sunday_missing)
    print("\n평일 결측치:\n", weekday_missing)

    
    # 결과 반환
    return saturday_missing, sunday_missing, weekday_missing

# 함수 실행
saturday_missing, sunday_missing, weekday_missing = check_missing_values_by_day(congestion_8_cleaned)
```

```python
import pandas as pd

# 1. 시간 컬럼을 'melt'로 행(Row) 데이터로 변환
melted_data = pd.melt(
    congestion_8_cleaned,
    id_vars=['요일구분', '호선', '역번호', '출발역', '상하구분'],
    value_vars=[col for col in congestion_8_cleaned.columns if '시' in col],
    var_name='시간',
    value_name='혼잡도'
)

# 2. "주말"로 묶기
melted_data.loc[melted_data['요일구분'].isin(['토요일', '일요일']), '요일구분'] = '주말'
melted_data.loc[~melted_data['요일구분'].isin(['주말']), '요일구분'] = '평일'

# 3. "출발역", "시간", "상하구분", "요일구분" 기준으로 혼잡도 평균 계산
result = (
    melted_data.groupby(['출발역', '시간', '상하구분', '요일구분'])
    .agg({'혼잡도': 'mean'})
    .reset_index()
)

# 4. 결과 확인
print(result)

# 5. 결과 저장 (선택)
result.to_csv('평일_주말_혼잡도_평균.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# 시간대별 혼잡도 평균 시각화
time_pattern = result.groupby(['시간', '요일구분'])['혼잡도'].mean().unstack()
time_pattern.plot(kind='line', figsize=(12, 6))
plt.title('시간대별 평일/주말 혼잡도 패턴')
plt.xlabel('시간')
plt.ylabel('평균 혼잡도')
plt.legend(title='요일구분')
plt.show()
```

```python
station_pattern = result.groupby(['출발역', '요일구분'])['혼잡도'].mean().unstack()
station_pattern.plot(kind='bar', figsize=(12, 6))
plt.title('역별 평일/주말 평균 혼잡도')
plt.xlabel('출발역')
plt.ylabel('평균 혼잡도')
plt.legend(title='요일구분')
plt.show()
```

```python
direction_pattern = result.groupby(['시간', '상하구분', '요일구분'])['혼잡도'].mean().unstack(['상하구분', '요일구분'])
direction_pattern.plot(kind='line', figsize=(12, 6))
plt.title('시간대별 상/하행선 혼잡도')
plt.xlabel('시간')
plt.ylabel('평균 혼잡도')
plt.legend(title='상하구분-요일구분')
plt.show()
```

```python
import pandas as pd

# 시간대별 데이터를 처리하는 함수
def process_and_aggregate_congestion(congestion_8):
    # 시간대 열 필터링 (예: '5시30분', '6시00분' 등)
    time_columns = [col for col in congestion_8.columns if '시' in col and '분' in col]
    
    # Melt 변환으로 시간대 데이터 구조화
    melted = congestion_8.melt(
        id_vars=['요일구분', '호선', '역번호', '출발역', '상하구분'],
        value_vars=time_columns,
        var_name='시간',
        value_name='혼잡도'
    )
    
    # 시간 열 변환 (예: '5시30분' → datetime.time 형식)
    melted['시간'] = pd.to_datetime(melted['시간'], format='%H시%M분', errors='coerce').dt.time
    
    # 요일별, 시간별 평균 혼잡도 계산
    time_agg = melted.groupby(['요일구분', '호선', '출발역', '시간', '상하구분'])['혼잡도'].mean().reset_index()
    
    # 컬럼 정리
    time_agg.rename(columns={
        '요일구분': '요일',
        '호선': '호선명',
        '출발역': '역명',
        '상하구분': '방향',
        '혼잡도': '평균혼잡도'
    }, inplace=True)
    
    return time_agg

# 시간대별 집계 실행
time_agg_congestion = process_and_aggregate_congestion(congestion_8)

# 결과 확인
print(time_agg_congestion)
```
